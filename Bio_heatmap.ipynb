```python
def zscore(proteins, row):
  intersect = pd.read_csv(f'intersect_{list(heatmap_df.index)[row]}.bed', header = None,
                          names = ['name', 'start_part', 'end_part', 'name2', 'z_score', 'name3', 'prom_start', 'prom_end', 'protein', 'strand','gen_prom_start', 'gen_end'], sep='\t')
  proteins = list(proteins.split(','))
  if proteins == ['*']:
    return -100000
  zscore = 0
  for protein in proteins:
    z_scores = list(intersect[intersect['protein'] == protein]['z_score'])
    if len(z_scores) == 0:
      continue
    zscore += sum(z_scores)/len(z_scores)

  if round(zscore/len(proteins),2) == 0:
    return -30000

  return round(zscore/len(proteins),2)
```

```python
# test
zscore('WP_159917519.1', 0)
```

```python
heatmap_df = clusters_df.iloc[:, 3:-1].T
for row in range(heatmap_df.shape[0]):
  for column in range(heatmap_df.shape[1]):
    heatmap_df[column][row] = zscore(heatmap_df[column][row], row)
heatmap_df
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

cmap = matplotlib.colors.LinearSegmentedColormap.from_list('', ['lightgrey', 'white', 'yellow', 'red'])

plt.figure(figsize=(20, 20))
sns.heatmap(heatmap_df.astype(float), annot=True, vmin=-100000, vmax=100000, cmap=cmap)
```
